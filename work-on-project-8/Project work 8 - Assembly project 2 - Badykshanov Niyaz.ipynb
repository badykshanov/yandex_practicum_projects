## Подготовка данных

### Технологический процесс, описание данных
Когда добытая руда проходит первичную обработку, получается дроблёная смесь. Её отправляют на флотацию (обогащение) и двухэтапную очистку. 
* Флотация

Во флотационную установку подаётся смесь золотосодержащей руды. После обогащения получается черновой концентрат и «отвальные хвосты», то есть остатки продукта с низкой концентрацией ценных металлов. *На стабильность этого процесса влияет непостоянное и неоптимальное физико-химическое состояние флотационной пульпы (смеси твёрдых частиц и жидкости).*

* Очистка

Черновой концентрат проходит две очистки. На выходе каждого получается концентрат и новые отвальные хвосты.

* Rougher feed — исходное сырье
* Rougher additions (или reagent additions) — флотационные реагенты: Xanthate, Sulphate, Depressant
* Xanthate — ксантогенат (промотер, или активатор флотации);
* Sulphate — сульфат (на данном производстве сульфид натрия);
* Depressant — депрессант (силикат натрия).
* Rougher process (англ. «грубый процесс») — флотация
* Rougher tails — отвальные хвосты
* Float banks — флотационная установка
* Cleaner process — очистка
* Rougher Au — черновой концентрат золота
* Final Au — финальный концентрат золота

*Параметры этапов*
* air amount — объём воздуха
* fluid levels — уровень жидкости
* feed size — размер гранул сырья
* feed rate — скорость подачи

*Наименование признаков*

Наименование признаков должно быть такое: [этап].[тип_параметра].[название_параметра]
Пример: rougher.input.feed_ag

*Возможные значения для блока [этап]:*
* rougher — флотация
* primary_cleaner — первичная очистка
* secondary_cleaner — вторичная очистка
* final — финальные характеристики

*Возможные значения для блока [тип_параметра]:*
* input — параметры сырья
* output — параметры продукта
* state — параметры, характеризующие текущее состояние этапа
* calculation — расчётные характеристики


### Метрика качества

Эффективность обогащения рассчитывается по формуле:

$Recovery = \frac{C \times (F-T)}{F \times (C-T))} \times 100\%$

где:
* C — доля золота в концентрате после флотации/очистки;
* F — доля золота в сырье/концентрате до флотации/очистки;
* T — доля золота в отвальных хвостах после флотации/очистки.

Для решения задачи используем метрику качества — sMAPE (англ. Symmetric Mean Absolute Percentage Error, «симметричное среднее абсолютное процентное отклонение»).

$sMAPE = \frac{1}{N} \sum_{i=1}^{N} \frac{| y_i - \hat{y}_i |}{(|y_i| + |\hat{y}_i|) / 2} \times 100\% $

где:
* $y_i$ - Значение целевого признака для объекта с порядковым номером i в выборке, на которой измеряется качество.
* $\hat{y}_i$ - Значение предсказания для объекта с порядковым номером i, например, в тестовой выборке.

Нужно спрогнозировать сразу две величины:
1.	эффективность обогащения чернового концентрата rougher.output.recovery;
2.	эффективность обогащения финального концентрата final.output.recovery.

Итоговая метрика складывается из двух величин:

$\begin{matrix}
  Итоговое \\
  sMAPE
 \end{matrix} = 25\% \times sMAPE(rougher) + 75\% \times sMAPE(final) $

### Проверка рассчета эффективности обогащения 

Вычислим эффективность обогащения на обучающей выборке для признака *rougher.output.recovery*. Найдем MAE между нашими расчётами и значением признака.

Исходя из формулы рассчета эффективности обогащения, 

$rougher.output.recovery = \frac{'rougher.output.concentrate\_au'\ \times \ ('rougher.input.feed\_au'\  - \ 'rougher.output.tail\_au')}                                  {'rougher.input.feed\_au'\  \times \ ('rougher.output.concentrate\_au'\  - \ 'rougher.output.tail\_au')} \  \times \  100\%$

Для *final.output.recovery*:

$final.output.recovery = \frac{'final.output.concentrate\_au'\ \times \ ('rougher.output.concentrate\_au'\  - \ 'final.output.tail\_au')}                                  {'rougher.output.concentrate\_au'\  \times \ ('final.output.concentrate\_au'\  - \ 'final.output.tail\_au')} \  \times \  100\%$

```python
# Для вычисления используем функцию
def 𝑟𝑒𝑐𝑜𝑣𝑒𝑟𝑦(C, F, T):
    """Calculates the answer according to its formula (C*(F-T)) / (F*(C-T)) * 100

    :param C: arg1
    :type C: float
    :param F: arg2
    :type F: float
    :param T: arg3а
    :type T: float
    
    :rtype: float
    :return: the result of the calculation according to the formula
    """
    R = (C*(F-T)) / (F*(C-T)) * 100
    return R

# Вычислим эффективность обогащения и сохраним в серии. Посмотрим на результат
predicted_recovery = train_df.apply(lambda row : 𝑟𝑒𝑐𝑜𝑣𝑒𝑟𝑦(row['rougher.output.concentrate_au'],
                     row['rougher.input.feed_au'], row['rougher.output.tail_au']), axis = 1)

predicted_recovery = pd.Series(predicted_recovery)
predicted_recovery
```

```python
# Для визуального сравнения посмотрим на метрику в обучающем наборе
train_df['rougher.output.recovery']
```

```python
# Найдем MAE между нашими расчётами и значением признака
print('MAE:', mean_absolute_error(predicted_recovery, train_df['rougher.output.recovery']))
```

## Построение моделей

### Функция для вычисления итоговой sMAPE

Напишем функцию для вычисления итоговой sMAPE

```python
# Напишем подфункцию, почти вычисляющую столбец с sMAPE, останется только найти среднее
def smape_elem(t_v, p_v, column):
    """Calculates the one column of Symmetric Mean Absolute Percentage Error

    :param t_v: DataFrame with prepared true data
    :type t_v: DataFrame
    :param p_v: DataFrame with prepared predicted data
    :type p_v: DataFrame
    :param column: name of column for calculate
    :type column: object
    
    :rtype: DataFrame
    :return: one column of Symmetric Mean Absolute Percentage Error
    """
    return (abs(t_v[column] - p_v[column]) / ((abs(t_v[column]) + abs(p_v[column])) / 2))
```

```python
# Функция для вычисления итоговой sMAPE
def smape(target_df, predictions_df):
    """Calculates aggregate Symmetric Mean Absolute Percentage Error

    :param target_df: DataFrame with true data
    :type target_df: DataFrame
    :param predictions_df: DataFrame with predicted data
    :type predictions_df: DataFrame
    
    :rtype: float
    :return: aggregate Symmetric Mean Absolute Percentage Error
    """
    target_df.reset_index(inplace = True, drop = True)
    target_df.columns = [0, 1]
    predictions_df = pd.DataFrame(predictions_df)
    rougher_smape = smape_elem(target_df, predictions_df, 1).mean()
    final_smape = smape_elem(target_df, predictions_df, 0).mean()
    symmetric_smape = 0.25 * rougher_smape + 0.75 * final_smape
    return symmetric_smape
```

### Обучение, проверка моделей

```python
# С помощью Pipeline и GridSearchCV найдем лучшие параметры для модели DecisionTreeRegressor
score = make_scorer(smape, greater_is_better=False)
cv_n = 5

pipe_DTRegressor = Pipeline([('scaler',  StandardScaler()),
            ('DTRegressor', DecisionTreeRegressor(random_state=SEED))])

grid_params_DTRegressor = [{'DTRegressor__max_depth': range (1, 13, 2),
                            'DTRegressor__min_samples_leaf': range (1, 8),
                            'DTRegressor__min_samples_split': range (2, 10, 2)
}]

CV_dtregressor = GridSearchCV (estimator = pipe_DTRegressor,
                               param_grid = grid_params_DTRegressor,
                               cv = cv_n,
                               scoring = score, verbose=0, return_train_score=True)

CV_dtregressor.fit(features, target)

dtr_best_params = CV_dtregressor.best_params_
dtr_best_model = CV_dtregressor.best_estimator_
dtr_score = CV_dtregressor.best_score_
```

```python
# Вычислим sMAPE модели к тестовой выборке
dtr_predictions_valid = dtr_best_model.predict(features_test)
print('Итоговое sMAPE дерева решений: {:.2%}'.format(smape(target_test, dtr_predictions_valid)))
```

```python
# Для RandomForestRegressor используем Pipeline и RandomizedSearchCV
pipe_RFRegressor = Pipeline([('scaler',  StandardScaler()),
            ('RFRegressor', RandomForestRegressor(random_state=SEED))])

grid_params_RFRegressor = [{'RFRegressor__n_estimators': range (10, 51, 10),
                    'RFRegressor__max_depth': range (1, 13, 2),
                    'RFRegressor__min_samples_leaf': range (1, 8),
                    'RFRegressor__min_samples_split': range (2, 10, 2)
}]

CV_rfregressor = RandomizedSearchCV (estimator = pipe_RFRegressor,
                               param_distributions = grid_params_RFRegressor,
                               cv = cv_n,
                               scoring = score, verbose=0, return_train_score=True)

CV_rfregressor.fit(features, target)

dfr_best_params = CV_rfregressor.best_params_
dfr_best_model = CV_rfregressor.best_estimator_
dfr_score = CV_rfregressor.best_score_
```

```python
# Вычислим sMAPE модели к тестовой выборке
dfr_predictions_valid = dfr_best_model.predict(features_test)
print('Итоговое sMAPE случайного леса: {:.2%}'.format(smape(target_test, dfr_predictions_valid)))
```

```python
# С помощью Pipeline и GridSearchCV найдем лучшие параметры для модели KNeighborsRegressor
score = make_scorer(smape, greater_is_better=False)
cv_n = 5

pipe_KNRegressor = Pipeline([('scaler',  StandardScaler()),
            ('KNRegressor', KNeighborsRegressor())])

grid_params_KNRegressor = [{'KNRegressor__n_neighbors': range (3, 11, 2),
                    'KNRegressor__weights': ('uniform', 'distance'),
                    'KNRegressor__algorithm': ('auto', 'ball_tree', 'kd_tree', 'brute')
}]

CV_knregressor = GridSearchCV (estimator = pipe_KNRegressor,
                               param_grid = grid_params_KNRegressor,
                               cv = cv_n,
                               scoring = score, verbose=0, return_train_score=True)

CV_knregressor.fit(features, target)

knr_best_params = CV_knregressor.best_params_
knr_best_model = CV_knregressor.best_estimator_
knr_score = CV_knregressor.best_score_
```

```python
# Вычислим sMAPE модели к тестовой выборке
knr_predictions_valid = knr_best_model.predict(features_test)
print('Итоговое sMAPE метода ближайшего соседа: {:.2%}'.format(smape(target_test, knr_predictions_valid)))
```

Лучшие результаты показала модель случайного леса. Посмотрим, какие метрики она считает самыми важными

```python
feature_names = features.columns

importances = CV_rfregressor.best_estimator_.named_steps["RFRegressor"].feature_importances_
std = np.std([
    tree.feature_importances_ for tree in CV_rfregressor.best_estimator_.named_steps["RFRegressor"].estimators_], axis=0)

forest_importances = pd.Series(importances, index=feature_names)

fig, ax = plt.subplots(figsize=(16,9))
forest_importances.plot.bar(yerr=std, ax=ax)
ax.set_title("Feature importances")
fig.tight_layout()
```

Здесь мы видим оценку значимости признаков. Для профессионалов некоторые данные могут быть интересными. Например, введение сульфатов на первом этапе очистки по мнению модели очень сильно влияет на эффективность обогащения. Впрочем, это и так наверняка им известно.

Проверим модели на адекватность, сравнив с моделями, которые предсказывают средние значения и медианы.
